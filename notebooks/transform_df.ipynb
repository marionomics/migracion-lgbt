# Transform Dataframe ENOE
Raw data from ENOE has to be transformed to be used. We use select variables to create our own indicators.
The files from ENOE are coded in this WAY:
* COE1 means it's the first survey.
* t1 means its from 1st quarter
* 10 means it's from the year 2010

therefore coe1t110.csv means that the data is from the first survey, first quarter and 2010.
